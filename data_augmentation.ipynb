# Data Augmentation

```python
import tensorflow
from tensorflow import keras
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/cat.jpeg', target_size=(250,250))
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)
```

```python
img_array = image.img_to_array(img)
```

```python
input_array = img_array.reshape((1,) + img_array.shape)
```

```python
i = 0

for batch in datagen.flow(input_array, batch_size=1, save_to_dir='/content/preview', save_prefix='cat', save_format='jpeg'):
  i += 1
  if i > 20:
    break
```

```python
#Now show all images from preview folder

import os
images = os.listdir('/content/preview')

#show images in serial manner all inside preview folder

for i in images:
  print(i)
  img = image.load_img('/content/preview/' + i, target_size=(250,250))
  display(img)


```
